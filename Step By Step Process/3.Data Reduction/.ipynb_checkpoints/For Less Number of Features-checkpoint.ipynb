### 2. Low Variance filter
Consider a variable in our dataset where all the observations have the same value, say 1. If we use this variable, do you think it can improve the model we will build? The answer is no, because this variable will have zero variance.

So, we need to calculate the variance of each variable we are given. Then drop the variables having low variance as compared to other variables in our dataset. The reason for doing this, as I mentioned above, is that variables with a low variance will not affect the target variable.

Steps for Low variance filter

```python
## Step 1.Check if there are any null values and check the data types of the columns
data = df.copy()
data.isnull().sum()
```

```python
data.dtypes
```

```python
## Step 2- Normalize the data using normalize module in sklearn.preprocessing

from sklearn.preprocessing import normalize
data_scaled = normalize(data)
data_scaled = pd.DataFrame(data_scaled,columns = data.columns)
data_scaled
```

```python
### step 4- Calculate variance using var() function
var = data_scaled.var()
var
```

```python
## Step 4 - Observe the variance and set a threshold for variance
threshold = 3/10000000
print(threshold)
```

```python
## Step 5 - Select the features which have variance greater than the threshold

data_var = pd.DataFrame({
    'columns':data.columns,
    'variance':var
})
data_var
```

```python
columns_selected = data_var[data_var['variance']>0.00000004]['columns'].values
```

```python
data_selected = data_scaled[columns_selected]
columns_selected
```

```python
data_selected.head()
```

### 3. High Correlation Filter or Multicollinearity
Multicollinearity occurs when two or more independent variables in a data frame have a high correlation with one another in a regression model. This means that one independent variable can be predicted from another in a regression model.

Let me take a simple **example** from our everyday life to explain this. Colin loves watching television while munching on chips. The more television he watches, the more chips he eats, and the happier he gets!

Now, if we could quantify happiness and measure Colin’s happiness while he’s busy doing his favorite activity, which do you think would have a greater impact on his happiness? Having chips or watching television? That’s difficult to determine because the moment we try to measure Colin’s happiness from eating chips, he starts watching television. And the moment we try to measure his happiness from watching television, he starts eating chips.

#### The Problem With Having Multicollinearity
Multicollinearity can be a problem in a regression model when using algorithms such as OLS (ordinary least squares) in statsmodels. This is because the estimated regression coefficients become unstable and difficult to interpret in the presence of multicollinearity. Statsmodels is a Python library that provides a range of tools for statistical analysis, including regression analysis.

### Y = W0+W1*X1+W2*X2

Coefficient W1 is the increase in Y for a unit increase in X1 while keeping X2 constant. But since X1 and X2 are highly correlated, changes in X1 would also cause changes in X2, and we would not be able to see their individual effect on Y.

### Detecting Multicollinearity Using a Variance Inflation Factor (VIF)

In Python, there are several ways to detect multicollinearity in a dataset, such as using the Variance Inflation Factor (VIF) or calculating the correlation matrix of the independent variables. To address multicollinearity, techniques such as regularization or feature selection can be applied to select a subset of independent variables that are not highly correlated with each other. In this article, we will focus on the most common one – VIF (Variance Inflation Factors).

**VIF determines the strength of the correlation between the independent variables. It is predicted by taking a variable and regressing it against every other variable.**

R^2 value is determined to find out how well an independent variable is described by the other independent variables. A high value of R^2 means that the variable is highly correlated with the other variables. This is captured by the VIF, which is denoted below:



So, the closer the R^2 value to 1, the higher the value of VIF and the higher the multicollinearity with the particular independent variable.

#### VIF starts at 1 and has no upper limit
#### VIF = 1, no correlation between the independent variable and the other variables
#### VIF exceeding 5 or 10 indicates high multicollinearity between this independent variable and the others

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
X = df.iloc[:,:-1]
calc_vif(X)
```

```python
sns.heatmap(X.corr(),annot=True)
```

From the VIF value and the heatmap we can see that tc and ldl is highly correlated.
### Fixing Multicollinearity
Dropping one of the correlated features will help in bringing down the multicollinearity between correlated features:

```python
X = X.drop(['tc'],axis=1)
calc_vif(X)
```

```python
sns.heatmap(X.corr(),annot=True)
```

```python
X = X.drop(['tch'],axis=1)
calc_vif(X)
```

```python
sns.heatmap(X.corr(),annot=True)
```

### 4. Backward Feature Elimination
Let us suppose we have the data which does not contain missing values, no features have low variance and there is no multicollinearity among the independent variables. In this case the backward feature elimination is the other way using which we can try to find features with lowest importance in building the model.

**Example**
Let us say we are trying to predict the final marks of a student using assignment1, assignment2 and assignment3.

- First the model is built using all the features (assignments) and model performance is evaluated
- Next, assignment1 is removed and model is built on assignment2 and assignment3, Model performance is evaluated
- Next, assignment2 is removed and model is built on assignment1 and assignment3, Model performance is evaluated
- Next, assignment3 is removed and model is built on assignment1 and assignment2, Model performance is evaluated
- After completing all the combinations of features, We will check the performance of the model with least change in performance when compared to initial model containing all the features.

We will identify the feature removed in the model and will not use the feature in our model build

**Following are the steps to perform backward feature elimination:**

1. Read the csv using pandas library and display the first few rows
2. Check if there are any null values and check the data types of the columns

```python
data_back = df.copy()
data_back
```

```python
data_back.isnull().sum()
```

3. Select X and y variables for building model, Dataframe X typically contains independent variables and y contains dependent variable (Drop the columns such as ID from independent variables since those do not help in building the model)

```python
X = data_back.iloc[:,:-1]
y = data_back['progression']
```

4. Import SequentialFeatureSelector from mlxtend.feature_selection module

5. Import required libraries for building the model, In this case we are using LinearRegression from sklearn.linear_model since we are trying to build a regression model

6. Create an object for LinearRegression and SequentialFeatureSelector, SequentialFeatureSelector takes following arguments as inputs:

    **Model :** In this case it is linear regression

    **k_features:** Number of features we want to have in the final model build

    **forward:** False (Since we are performing backward feature elimination)

    **verbose:** 1(displays execution messages)

    **scoring:** Evaluation metric for the model (In this case we are using “negative mean squared error”

7. Perform fit for SequentialFeatureSe

```python
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
sfs_initial = sfs(lin_reg,k_features = 5,forward = False,verbose = 1,scoring = 'neg_mean_squared_error')
## In this case we are extracting 5 important features as k_features = 5.
```

```python
sfs_initial.fit(X,y)
```

8. Identify the features selected using k_feature_names_ variable property of sfs_initial object

9. Use the features selected by the SequentialFeatureSelector for final model build

```python
vars_selected = list(sfs_initial.k_feature_names_)
updated_data = data_back[vars_selected]
updated_data
```
