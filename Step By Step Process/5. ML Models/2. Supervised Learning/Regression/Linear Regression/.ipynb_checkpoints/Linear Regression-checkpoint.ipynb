# What is Linear Regression?
Linear regression is a type of statistical analysis used to predict the relationship between two variables. It assumes a linear relationship between the independent variable and the dependent variable, and aims to find the best-fitting line that describes the relationship. The line is determined by minimizing the sum of the squared differences between the predicted values and the actual values.

Linear regression is commonly used in many fields, including economics, finance, and social sciences, to analyze and predict trends in data. It can also be extended to multiple linear regression, where there are multiple independent variables, and logistic regression, which is used for binary classification problems.

## Simple Linear Regression
In a simple linear regression, there is one independent variable and one dependent variable. The model estimates the slope and intercept of the line of best fit, which represents the relationship between the variables. The **slope** represents the change in the dependent variable for each unit change in the independent variable, while the **intercept** represents the predicted value of the dependent variable when the independent variable is zero.

Linear regression is a quiet and the simplest statistical regression method used for predictive analysis in machine learning. Linear regression shows the linear relationship between the independent(predictor) variable i.e. X-axis and the dependent(output) variable i.e. Y-axis, called linear regression. If there is a single input variable X(independent variable), such linear regression is called simple linear regression.



**The above graph presents the linear relationship between the output(y) variable and predictor(X) variables.  The blue line is referred to as the best fit straight line. Based on the given data points, we attempt to plot a line that fits the points the best.**

To calculate best-fit line linear regression uses a traditional slope-intercept form which is given below,

**Yi = 0 + 1Xi**

where,

##### Yi = Dependent variable,
##### 0 = constant/Intercept,
##### 1 = Slope/Intercept, 
##### Xi = Independent variable.

##### Exploring 1

- If 1 > 0, then x(predictor) and y(target) have a positive relationship. That is increase in x will increase y.
- If 1 < 0, then x(predictor) and y(target) have a negative relationship. That is increase in x will decrease y.

##### Exploring 0

- If the model does not include x=0, then the prediction will become meaningless with only b0. For example, we have a dataset that relates height(x) and weight(y). Taking x=0(that is height as 0), will make equation have only b0 value which is completely meaningless as in real-time height and weight can never be zero. This resulted due to considering the model values beyond its scope.
- If the model includes value 0, then b0 will be the average of all predicted values when x=0. But, setting zero for all the predictor variables is often impossible.
- The value of b0 guarantee that residual have mean zero. If there is no b0 term, then regression will be forced to pass over the origin. Both the regression co-efficient and prediction will be biased.


This algorithm explains the linear relationship between the dependent(output) variable y and the independent(predictor) variable X using a straight line  Y= 0 + 1 X.



**The goal of the linear regression algorithm is to get the best values for B0 and B1 to find the best fit line. The best fit line is a line that has the least error which means the error between predicted values and actual values should be minimum.**

### How to Calculate Slope and Intercept in linear Regression
To calculate the intercept and slope in linear regression, you can use the method of least squares. The least squares method minimizes the sum of the squared differences between the observed dependent variable values and the predicted values based on the linear regression equation.

Here's a step-by-step guide on how to calculate the intercept and slope:

1. **Gather your data:** Collect the data for the independent variable (X) and the dependent variable (Y). Ensure that you have a sufficient number of data points to perform the regression analysis.

2. **Calculate the means:** Calculate the means (average values) of the X and Y variables. Let's denote the means as X and , respectively.

3. **Calculate the deviations:** Calculate the deviation of each X value from the mean of X i.e. (X - X) and the deviation of each Y value from the mean of Y i.e. (Y - ).

4. **Calculate the sums:** Calculate the sum of the product of the deviations (X - X) * (Y - ), as well as the sum of the squared deviations of X, (X - X).

5. **Calculate the slope ():** The slope of the linear regression line can be calculated using the following formula:

__ = [(X - X) * (Y - )] / [(X - X)]__

6. **Calculate the intercept ():** Once you have the slope, you can calculate the intercept using the following formula:

** =  -  * X**

7. **Interpret the results:** The calculated slope () represents the change in the dependent variable (Y) for each unit increase in the independent variable (X). The intercept () represents the value of the dependent variable (Y) when the independent variable (X) is zero.

By following these steps, you can calculate the intercept and slope for a linear regression analysis.

## Cost Function
In linear regression, the cost function, also known as the loss function or the objective function, is a mathematical function that measures the difference between the predicted values and the actual values of the dependent variable in a dataset. The goal of linear regression is to find the best-fit line that minimizes this cost function.

The most commonly used cost function in linear regression is the mean squared error (MSE) function. It calculates the average squared difference between the predicted values and the actual values. The formula for MSE is:

### MSE = (1/n) * (y - )

where:

- **MSE** is the mean squared error
- **n** is the number of data points
- **y** is the actual value of the dependent variable for the i-th data point
- **** is the predicted value of the dependent variable for the i-th data point

The cost function quantifies the quality of the model's predictions by assigning a numerical value to the error between the predicted and actual values. By minimizing the cost function, the linear regression algorithm adjusts the model's parameters, such as the slope and intercept of the line, to find the line that best fits the data.

The uses of the cost function in linear regression include:

1. **Model training:** During the training process, the cost function is minimized using optimization algorithms like gradient descent. By iteratively adjusting the model's parameters based on the cost function, the algorithm finds the optimal line that minimizes the overall prediction error.
2. **Evaluation of model performance:** The cost function provides a measure of how well the linear regression model fits the training data. Lower values of the cost function indicate better model performance, as they indicate smaller prediction errors.
3. **Comparison of different models:** The cost function allows for the comparison of different models or variations of linear regression. By comparing the cost function values of different models, you can determine which model performs better in terms of minimizing prediction errors.
4. **Feature selection:** The cost function can be used to evaluate the importance or relevance of different features in the linear regression model. By assessing how the cost function changes when specific features are included or excluded, you can determine which features contribute the most to the model's predictive power.

**Overall, the cost function is a crucial component of linear regression, guiding the model's learning process and assessing its performance.**

## Gradient descent
Gradient descent is an optimization algorithm commonly used in linear regression to minimize the cost function and find the optimal parameters for the linear regression model. The goal of gradient descent is to iteratively update the model's parameters in the direction of steepest descent of the cost function until a minimum is reached.

Here's how gradient descent works in the context of linear regression:

1. **Initialization:** Start by initializing the model's parameters, typically the slope (coefficients) and intercept (bias) of the linear regression line, with random values or zeros.

2. **Calculate the cost function:** Evaluate the cost function (such as mean squared error) using the current parameter values and the training data. This measures the prediction error of the current model.

3. **Compute gradients:** Calculate the gradients of the cost function with respect to each parameter. The gradient represents the direction and magnitude of the steepest ascent of the cost function. In linear regression, the gradients can be calculated analytically using calculus.

4. **Update parameters:** Adjust the parameters by taking a small step in the opposite direction of the gradients. The step size is determined by the learning rate, which controls the size of the parameter updates. The update rule for each parameter is:

##  :=  - learning_rate * gradient

where **** is the i-th parameter, learning_rate is the learning rate hyperparameter, and gradient is the gradient of the cost function with respect to .

5. **Repeat steps 2-4:** Iterate the process of calculating the cost function, computing gradients, and updating parameters until a termination condition is met. This condition could be a maximum number of iterations, reaching a desired level of convergence, or a small change in the cost function between iterations.

6. **Final parameter values:** Once the algorithm terminates, the final parameter values represent the optimal values that minimize the cost function, resulting in the best-fit line for the linear regression model.

By iteratively adjusting the parameters in the direction of the steepest descent of the cost function, gradient descent gradually converges to the optimal solution. The learning rate determines the step size taken at each iteration, balancing the trade-off between convergence speed and stability.

It's worth noting that there are variations of gradient descent, such as stochastic gradient descent (SGD) and mini-batch gradient descent, which introduce randomness or use smaller subsets of the training data to update the parameters. These variations can be useful for large datasets to improve computational efficiency and prevent getting stuck in local optima.

## Evaluation Metrics for Linear Regression

When evaluating the performance of a linear regression model, several evaluation metrics can be used to assess its accuracy and effectiveness in predicting the dependent variable. Here are some commonly used evaluation metrics for linear regression:

1. **Mean Squared Error (MSE):** MSE calculates the average squared difference between the predicted values and the actual values. It provides a measure of the overall prediction error of the model. Lower MSE values indicate better model performance.

### MSE = (1/n) * (y - )

- **y** is the actual value of the dependent variable for the i-th data point

3. **Mean Absolute Error (MAE):** MAE calculates the average absolute difference between the predicted values and the actual values. It provides a measure of the average magnitude of the prediction error. Lower MAE values indicate better model performance.

5. **Adjusted R-squared:** Adjusted R-squared adjusts for the number of predictors in the model. ## Assumptions of Linear Regression
Linear regression relies on several key assumptions to ensure the validity and accuracy of the model's estimates and predictions. These assumptions are as follows:

1. **Linearity:** The relationship between the independent variables and the dependent variable is assumed to be linear. The model assumes that the effect of the independent variables on the dependent variable is additive and constant.

2. **Independence:** The observations in the dataset are assumed to be independent of each other. This assumption ensures that the errors or residuals are not correlated and do not exhibit any systematic patterns.

3. **Homoscedasticity:** The variance of the errors or residuals is assumed to be constant across all levels of the independent variables. In other words, the spread of the residuals should be consistent across the range of predicted values. Homoscedasticity ensures that the model's predictions are equally accurate and precise across the entire range of the dependent variable.




4. **Normality:** The errors or residuals are assumed to follow a normal distribution. This assumption enables statistical inference, hypothesis testing, and confidence interval estimation. It also ensures that the maximum likelihood estimates of the model's coefficients are unbiased and efficient.

5. **No multicollinearity:** The independent variables should not exhibit strong correlations or multicollinearity among themselves. Multicollinearity can cause instability in the estimated coefficients and make it challenging to interpret the individual effects of the independent variables.

6. **No endogeneity:** The independent variables are assumed to be exogenous and not influenced by the errors or residuals. Endogeneity occurs when there is a two-way causal relationship between the dependent variable and one or more of the independent variables. Violation of this assumption can lead to biased and inconsistent coefficient estimates.

7. **No influential outliers:** The presence of influential outliers can disproportionately influence the regression results and distort the estimated coefficients. It is important to identify and handle outliers appropriately to ensure the validity of the model.

These assumptions provide the foundation for reliable and valid inference in linear regression analysis. Violation of these assumptions can lead to biased estimates, unreliable predictions, and incorrect interpretation of the results. It is crucial to assess these assumptions and, if necessary, employ appropriate techniques to address violations or consider alternative modeling approaches.

## Splitting of Data in Test and Training Dataset using train_test_split

The train_test_split function in scikit-learn is a utility function that helps in splitting a dataset into training and testing sets. It is commonly used in machine learning tasks to evaluate the performance of models on unseen data. The function splits the data randomly by default, but you can control the splitting behavior using various parameters. Here are the important parameters of train_test_split:

1. **arrays:** This parameter represents the data arrays that you want to split into training and testing sets. It could be a single array or multiple arrays.

2. **test_size:** This parameter specifies the proportion of the data that should be allocated to the testing set. It can be set as a float value between 0 and 1, indicating the fraction of the data (e.g., 0.2 for 20%) or as an integer representing the absolute number of samples to allocate for testing. The default value is 0.25, meaning 25% of the data will be used for testing.

3. **train_size:** This parameter specifies the proportion of the data that should be allocated to the training set. It works in the same way as test_size. If train_size is not provided, it will be automatically set to the complement of test_size, i.e., 1 - test_size.

4. **random_state:** This parameter allows you to control the random shuffling of the data before splitting. It can take an integer value or an instance of the RandomState class. Setting a specific random_state value ensures reproducibility of the same split each time you run the code. If you don't provide a random_state value, the data will be randomly shuffled each time the code is executed.The random_state parameter serves two primary purposes:

    - **Reproducibility:** Setting a specific value for random_state ensures that the random splitting of data is reproducible. By using the same random_state value, you can obtain the same split every time you run your code. This is useful when you want to compare different models or methods and need consistent splits for fair comparisons.

    - **Randomness control:** Random shuffling is performed by train_test_split to create random training and testing sets. By specifying a random_state value, you can control the randomness and obtain a fixed random split. If you don't provide a random_state value, a different random split will be generated each time you run the code.

5. **shuffle:** This parameter controls whether the data should be shuffled before splitting. By default, it is set to True, which means the data will be randomly shuffled. If you want to disable shuffling, you can set shuffle=False.

6. **stratify:** This parameter is used for stratified sampling. If your dataset contains a target variable with class labels, you can set stratify to that variable. The data will be split in a way that maintains the same class distribution in both the training and testing sets.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, train_size = 0.7, test_size = 0.3, random_state = 100 )
```
